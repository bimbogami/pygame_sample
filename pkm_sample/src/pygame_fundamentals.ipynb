## 1. What is Pygame?

**Pygame** is a free, open-source Python library used for making **2D games and multimedia applications**. It is built on top of the **SDL (Simple DirectMedia Layer)** library and provides modules for:

| Module | Purpose |
|---|---|
| `pygame.display` | Managing the game window |
| `pygame.draw` | Drawing shapes (rect, circle, line, etc.) |
| `pygame.image` | Loading and saving images |
| `pygame.event` | Handling keyboard, mouse, and other events |
| `pygame.font` | Rendering text |
| `pygame.mixer` | Playing sounds and music |
| `pygame.time` | Managing time and frame rate |
| `pygame.transform` | Scaling, rotating, and flipping images |

Pygame is **beginner-friendly** and perfect for learning game development concepts like game loops, collision detection, and sprite animation.

## 4. Creating a Display Window

The **display window** (also called the **screen** or **surface**) is the canvas where everything is drawn.

### Key Functions

| Function | Description |
|---|---|
| `pygame.display.set_mode((width, height))` | Creates a window with the given dimensions |
| `pygame.display.set_caption("title")` | Sets the window title bar text |
| `pygame.display.flip()` | Updates the **entire** screen |
| `pygame.display.update()` | Updates only a portion (or all) of the screen |

### How It Works

Pygame uses a **double-buffering** technique:
1. You draw everything onto the screen **surface** (the back buffer).
2. You call `pygame.display.flip()` to swap the back buffer to the front, making everything visible at once.

This prevents **flickering**.

````python
import pygame
pygame.init()

# Define screen size / Created variables for screen height and width to be called in the screen or as default resolution
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600

# Create the display window
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("My First Pygame Window")

# Fill the screen with a color and display it
screen.fill((30, 30, 60))  # Dark blue-gray background
pygame.display.flip()

# Keep the window open for 3 seconds so you can see it
pygame.time.wait(3000)
pygame.quit() #closes after 3 second mark 
print("Window closed.")
````

## 6. Drawing Shapes

Pygame provides the `pygame.draw` module to draw basic shapes onto a surface.

### Common Drawing Functions

| Function | Description |
|---|---|
| `pygame.draw.rect(surface, color, rect)` | Draws a rectangle. `rect` = `(x, y, width, height)` |
| `pygame.draw.circle(surface, color, center, radius)` | Draws a circle |
| `pygame.draw.line(surface, color, start, end, width)` | Draws a straight line |
| `pygame.draw.ellipse(surface, color, rect)` | Draws an ellipse inside a bounding rect |
| `pygame.draw.polygon(surface, color, points)` | Draws a polygon from a list of points |

### Filled vs. Outlined

- **Filled shape**: omit the `width` parameter or set it to `0`.
- **Outlined shape**: pass a positive `width` value (thickness in pixels).

````python
import pygame
pygame.init()

screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Drawing Shapes")
screen.fill((30, 30, 60))

# --- Draw a filled rectangle ---
# pygame.draw.rect(surface, color, (x, y, width, height))
pygame.draw.rect(screen, (255, 100, 100), (50, 50, 200, 100))

# --- Draw an outlined rectangle ---
pygame.draw.rect(screen, (100, 255, 100), (300, 50, 200, 100), 3)  # 3px border

# --- Draw a filled circle ---
# pygame.draw.circle(surface, color, (center_x, center_y), radius)
pygame.draw.circle(screen, (100, 100, 255), (150, 300), 80)

# --- Draw a line ---
# pygame.draw.line(surface, color, (start_x, start_y), (end_x, end_y), width)
pygame.draw.line(screen, (255, 255, 0), (350, 250), (700, 400), 4)

# --- Draw a polygon (triangle) ---
pygame.draw.polygon(screen, (255, 0, 255), [(600, 100), (700, 250), (500, 250)])

pygame.display.flip()
pygame.time.wait(4000)
pygame.quit()
print("Shapes demo closed.")
````

## 7. The Game Loop

The **game loop** is the heart of every Pygame application. It runs continuously and performs three essential steps each frame:

```
┌──────────────────────────────────────┐
│           GAME LOOP (while)          │
│                                      │
│  1. HANDLE EVENTS  (input)           │
│     └─ keyboard, mouse, quit, etc.   │
│                                      │
│  2. UPDATE STATE   (logic)           │
│     └─ move objects, check collisions│
│                                      │
│  3. RENDER / DRAW  (output)          │
│     └─ clear screen, draw, flip      │
│                                      │
└──────────────────────────────────────┘
```

### Why do we need a loop?

Without a loop, the window would **appear and immediately close**. The loop keeps the window alive and responsive.

### Basic Structure

````python
import pygame
pygame.init()

screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Basic Game Loop")
clock = pygame.time.Clock()

running = True
while running:
    # ─── 1. EVENT HANDLING ───
    for event in pygame.event.get():
        if event.type == pygame.QUIT:      # User clicked the X button
            running = False

    # ─── 2. UPDATE (game logic) ───
    # (nothing to update in this basic example)

    # ─── 3. DRAW / RENDER ───
    screen.fill((30, 30, 60))              # Clear the screen each frame
    pygame.draw.circle(screen, (0, 200, 255), (400, 300), 50)
    pygame.display.flip()                  # Show everything

    clock.tick(60)                         # Cap at 60 FPS

pygame.quit()
print("Game loop ended.")
````

## 8. Event Handling

Pygame collects all user input (keyboard, mouse, window events) into an **event queue**. You read the queue each frame with `pygame.event.get()`.

### Common Event Types

| Event Type | Description |
|---|---|
| `pygame.QUIT` | User closed the window |
| `pygame.KEYDOWN` | A key was **pressed** down |
| `pygame.KEYUP` | A key was **released** |
| `pygame.MOUSEBUTTONDOWN` | A mouse button was clicked |
| `pygame.MOUSEMOTION` | The mouse was moved |

### Keyboard Input

When a `KEYDOWN` event occurs, you can check **which** key was pressed via `event.key`:

```python
if event.type == pygame.KEYDOWN:
    if event.key == pygame.K_SPACE:
        print("Space bar pressed!")
    elif event.key == pygame.K_ESCAPE:
        running = False
```

### Continuous Key Press

For **smooth, continuous movement** (e.g., moving a character), use `pygame.key.get_pressed()` instead of events:

```python
keys = pygame.key.get_pressed()
if keys[pygame.K_LEFT]:
    x -= 5
if keys[pygame.K_RIGHT]:
    x += 5
```

````python
import pygame
pygame.init()

screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Event Handling Demo")
clock = pygame.time.Clock()

# Circle position (will be moved by arrow keys)
x, y = 400, 300
speed = 5
color = (0, 200, 255)

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        # Single key press event — change color on SPACE
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                # Toggle between cyan and red
                color = (255, 80, 80) if color == (0, 200, 255) else (0, 200, 255)

    # Continuous key press — smooth movement with arrow keys
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT]:
        x -= speed
    if keys[pygame.K_RIGHT]:
        x += speed
    if keys[pygame.K_UP]:
        y -= speed
    if keys[pygame.K_DOWN]:
        y += speed

    # Draw
    screen.fill((30, 30, 60))
    pygame.draw.circle(screen, color, (x, y), 30)
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
print("Event demo closed.")
````

## 9. Loading and Displaying Images

To display images (sprites, backgrounds, etc.), use `pygame.image.load()`.

### Key Functions

| Function | Description |
|---|---|
| `pygame.image.load("path.png")` | Loads an image file into a Surface |
| `image.convert()` | Converts to display format (faster blitting) |
| `image.convert_alpha()` | Like `convert()` but preserves transparency |
| `screen.blit(image, (x, y))` | Draws the image onto the screen at position `(x, y)` |
| `pygame.transform.scale(image, (w, h))` | Resizes an image to `(width, height)` |

### What is "blitting"?

**Blit** (Block Image Transfer) means copying pixel data from one surface to another. When you call `screen.blit(image, (x, y))`, you're pasting the image onto the screen surface at that position.

### Transparency / Color Keys

If your sprite has a solid background color you want to make transparent, use `set_colorkey()`:

```python
image = pygame.image.load("sprite.png").convert()
image.set_colorkey((112, 154, 209))  # Makes this specific color transparent
```

````python
import pygame
pygame.init()

screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Image Loading Demo")

# --- Loading an image ---
# Replace with your actual image path
try:
    image = pygame.image.load("src/assets/sprites/rayquaza/idle/rayquaza_0.png").convert_alpha()
    
    # Scale the image to 2x its original size
    image = pygame.transform.scale(image, (image.get_width() * 2, image.get_height() * 2))
    
    # Remove background color (color key)
    image.set_colorkey((112, 154, 209))
    
    screen.fill((30, 30, 60))
    screen.blit(image, (300, 200))  # Draw image at position (300, 200)
    pygame.display.flip()
    pygame.time.wait(4000)
    
except FileNotFoundError:
    print("Image file not found! Make sure the path is correct.")
    print("Expected: src/assets/sprites/rayquaza/idle/rayquaza_0.png")

pygame.quit()
print("Image demo closed.")
````

## 10. Rendering Text

To display text on screen, use the `pygame.font` module.

### Steps to Render Text

1. **Create a Font object** — choose a font and size.
2. **Render the text** — this creates a new Surface with the text drawn on it.
3. **Blit the text surface** onto the screen.

```python
font = pygame.font.SysFont("Arial", 36)           # System font, size 36
text_surface = font.render("Hello!", True, WHITE)  # (text, antialias, color)
screen.blit(text_surface, (x, y))                  # Draw it
```

| Function | Description |
|---|---|
| `pygame.font.SysFont(name, size)` | Creates a font from a system font |
| `pygame.font.Font(file, size)` | Creates a font from a `.ttf` file |
| `font.render(text, antialias, color)` | Renders text into a Surface |

````python
import pygame
pygame.init()

screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Text Rendering Demo")
screen.fill((30, 30, 60))

# Create font objects
title_font = pygame.font.SysFont("Arial", 48, bold=True)
body_font  = pygame.font.SysFont("Arial", 28)

# Render text surfaces
title = title_font.render("Pygame Text Demo", True, (255, 255, 255))
line1 = body_font.render("This is rendered using pygame.font", True, (0, 200, 255))
line2 = body_font.render("Press any key or close the window.", True, (200, 200, 200))

# Blit text onto the screen
screen.blit(title, (200, 150))
screen.blit(line1, (180, 250))
screen.blit(line2, (180, 300))

pygame.display.flip()
pygame.time.wait(4000)
pygame.quit()
print("Text demo closed.")
````

## 11. Clock and Frame Rate

Without frame rate control, your game loop would run as fast as the CPU allows — making the game **too fast** and **inconsistent** across different machines.

### `pygame.time.Clock()`

- Create a `Clock` object **once** before the loop.
- Call `clock.tick(fps)` at the **end** of each frame to limit the frame rate.

```python
clock = pygame.time.Clock()

while running:
    # ... game logic and drawing ...
    clock.tick(60)  # Limits to 60 frames per second
```

### Common FPS Values

| FPS | Use Case |
|---|---|
| 30 | Simple games, turn-based games |
| 60 | Standard for most games |
| 14 | Sprite animation (e.g., Pokémon-style) |

> **Tip:** In this project's `main.py`, we use `clock.tick(14)` to match the frame rate to the sprite animation speed.

````python
import pygame
pygame.init()

screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("FPS Demo")
clock = pygame.time.Clock()
font = pygame.font.SysFont("Arial", 28)

x = 0  # Moving rectangle position

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Move a rectangle across the screen
    x += 3
    if x > 800:
        x = -100

    screen.fill((30, 30, 60))
    pygame.draw.rect(screen, (0, 200, 255), (x, 260, 100, 80))

    # Show the current FPS on screen
    fps_text = font.render(f"FPS: {clock.get_fps():.1f}", True, (255, 255, 255))
    screen.blit(fps_text, (10, 10))

    pygame.display.flip()
    clock.tick(60)  # Try changing this to 30 or 10 and see the difference!

pygame.quit()
print("FPS demo closed.")
````

## 13. Sprite Animation

Animation in Pygame works by **cycling through a list of images** (frames) over time. Each frame of the game loop displays the next image in the sequence.

### How Animation Works

```
Frame 0      Frame 1      Frame 2      Frame 3      ...
┌──────┐    ┌──────┐    ┌──────┐    ┌──────┐
│ 🟢 │    │  🟢  │    │   🟢 │    │    🟢│    → Loop back to Frame 0
└──────┘    └──────┘    └──────┘    └──────┘
```

### Steps

1. **Load all frames** into a list.
2. Keep a **frame index** variable.
3. Each game loop iteration, draw the current frame and **increment** the index.
4. When the index exceeds the number of frames, **reset to 0** (loop).

### Animation States

You can have multiple animation states (e.g., `"idle"` and `"act"`) by using separate frame lists and switching between them:

```python
if self.state == "act":
    image = self.act_frames[self.frame_index]
else:
    image = self.idle_frames[self.frame_index]
```

### Example: Simple Animation Class

Here's a simplified version of the animation pattern used in this project's `animation.py`:

````python
import pygame
pygame.init()

class SimpleAnimator:
    """A simple sprite animator that cycles through a list of images."""

    def __init__(self, frame_paths, scale=2):
        # Load all frame images into a list
        self.frames = []
        for path in frame_paths:
            img = pygame.image.load(path).convert_alpha()
            # Scale the image
            img = pygame.transform.scale(img, (img.get_width() * scale, img.get_height() * scale))
            # Remove background color
            img.set_colorkey((112, 154, 209))
            self.frames.append(img)

        self.current_frame = 0

    def update(self):
        """Advance to the next frame."""
        self.current_frame += 1
        if self.current_frame >= len(self.frames):
            self.current_frame = 0  # Loop back to the start

    def draw(self, screen, x, y):
        """Draw the current frame on the screen."""
        screen.blit(self.frames[self.current_frame], (x, y))


# --- Usage ---
screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Sprite Animation Demo")
clock = pygame.time.Clock()

# Build a list of frame file paths
# (Adjust these paths to match your actual sprite files)
try:
    frames = [f"assets/sprites/rayquaza/idle/rayquaza_{i}.png" for i in range(11)]
    animator = SimpleAnimator(frames)

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        animator.update()  # Go to next frame

        screen.fill((30, 30, 60))
        animator.draw(screen, 300, 200)
        pygame.display.flip()
        clock.tick(14)  # 14 FPS for a classic sprite animation feel

except FileNotFoundError as e:
    print(f"Sprite file not found: {e}")
    print("Make sure your sprite files exist in the correct path.")

pygame.quit()
print("Animation demo closed.")
````

## 14. Full Example: Putting It All Together

Below is a **complete Pygame application** that combines all the concepts covered in this tutorial:

- ✅ Initialization and display setup
- ✅ Color constants
- ✅ Game loop with event handling
- ✅ Drawing shapes
- ✅ Rendering text
- ✅ Clock and FPS control
- ✅ Keyboard input (arrow keys to move, SPACE to change color)

This example creates a **moving circle** that you can control with arrow keys and change its color with the space bar.

````python
import pygame
import sys

# ──────────────────────────────────────────────
# 1. INITIALIZATION
# ──────────────────────────────────────────────
pygame.init()

# Screen setup
SCREEN_WIDTH  = 800
SCREEN_HEIGHT = 600
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Pygame Fundamentals — Full Demo")

# Clock for frame rate control
clock = pygame.time.Clock()
FPS = 60

# ──────────────────────────────────────────────
# 2. COLORS
# ──────────────────────────────────────────────
BG_COLOR   = (30, 30, 60)
WHITE      = (255, 255, 255)
COLORS     = [
    (0, 200, 255),   # Cyan
    (255, 80, 80),   # Red
    (80, 255, 80),   # Green
    (255, 200, 50),  # Gold
    (200, 80, 255),  # Purple
]

# ──────────────────────────────────────────────
# 3. GAME STATE
# ──────────────────────────────────────────────
circle_x     = SCREEN_WIDTH // 2
circle_y     = SCREEN_HEIGHT // 2
circle_r     = 30
speed        = 5
color_index  = 0

# Font for HUD text
font = pygame.font.SysFont("Arial", 22)

# ──────────────────────────────────────────────
# 4. GAME LOOP
# ──────────────────────────────────────────────
running = True
while running:

    # ─── EVENT HANDLING ───
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                # Cycle through colors
                color_index = (color_index + 1) % len(COLORS)
            if event.key == pygame.K_ESCAPE:
                running = False

    # ─── CONTINUOUS INPUT (smooth movement) ───
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT]:
        circle_x -= speed
    if keys[pygame.K_RIGHT]:
        circle_x += speed
    if keys[pygame.K_UP]:
        circle_y -= speed
    if keys[pygame.K_DOWN]:
        circle_y += speed

    # Keep the circle on screen
    circle_x = max(circle_r, min(SCREEN_WIDTH - circle_r, circle_x))
    circle_y = max(circle_r, min(SCREEN_HEIGHT - circle_r, circle_y))

    # ─── DRAWING ───
    screen.fill(BG_COLOR)

    # Draw the player circle
    current_color = COLORS[color_index]
    pygame.draw.circle(screen, current_color, (circle_x, circle_y), circle_r)

    # Draw a UI panel at the bottom
    pygame.draw.rect(screen, (50, 50, 80), (0, SCREEN_HEIGHT - 60, SCREEN_WIDTH, 60))

    # Render HUD text
    hud = font.render(
        f"Position: ({circle_x}, {circle_y})  |  Color: {color_index+1}/{len(COLORS)}  |  FPS: {clock.get_fps():.0f}",
        True, WHITE
    )
    screen.blit(hud, (20, SCREEN_HEIGHT - 42))

    instructions = font.render("Arrow Keys = Move  |  Space = Change Color  |  Esc = Quit", True, (180, 180, 180))
    screen.blit(instructions, (20, 15))

    # ─── FLIP & TICK ───
    pygame.display.flip()
    clock.tick(FPS)

# ──────────────────────────────────────────────
# 5. CLEANUP
# ──────────────────────────────────────────────
pygame.quit()
print("Full demo ended. Great job completing the tutorial!")
````
